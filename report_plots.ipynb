```python
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
```

```python
models = ['NickNetV2', 'NickNetV2-SGD', 'MinimalErrPDetector']
auc_scores = [0.60808, 0.62541, 0.66125]
datasets = ['DefaultA', 'ensemble', 'Default']  # Which dataset each model was trained on

colors = {'Default': 'green', 'A': 'blue', 'B': 'yellow', 'C': 'red', 'DefaultA': 'orange', 'ensemble': 'purple'}  # Color mapping for datasets
names_map = {'Default': 'Default Dataset', 'A': 'Dataset A', 'B': 'Dataset B', 'C': 'Dataset C', 'DefaultA': 'Default-A', 'ensemble': 'ensemble'}
# Create the figure and axis
fig, ax = plt.subplots(figsize=(8, 6))
#plt.subplots_adjust(right=0.75) 
# Create bars with colors based on dataset
bars = []
for model, score, dataset in zip(models, auc_scores, datasets):
    bar = ax.bar(model, score, color=colors[dataset])
    bars.append(bar)
    
    # Add the AUC score text above each bar
    ax.text(model, score + 0.01, f'{score:.5f}', 
            ha='center', va='bottom', fontsize=12)

# Set y-axis limits and labels
ax.set_ylim(0, 1)
ax.set_ylabel('AUC Score', fontsize=12)
ax.set_xlabel('Models', fontsize=12)
ax.set_title('AUC Scores by Model of set B', fontsize=14, pad=20)

# Create legend for datasets
#legend_elements = [Patch(facecolor=color, label=names_map[dataset]) 
#                   for dataset, color in colors.items()]
legend_elements = [Patch(facecolor=colors[d], label=names_map[d]) for d in datasets]

ax.legend(handles=legend_elements, title='Training Dataset') #,
          #bbox_to_anchor=(1.05, 1),  
          #loc='upper left',           
          #borderaxespad=0.)

# Optional: Add grid lines for better readability
ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.savefig("docs/images/auc_set_b.png")
plt.show()
```
